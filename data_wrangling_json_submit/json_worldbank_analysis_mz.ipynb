```python
import numpy as np
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#read in data 
projects_df=pd.read_json('data/world_bank_projects.json')
#get an idea of your data
pd.set_option('display.max_colwidth', -1)
projects_df.head(20)
```

```python
#view the full row to understand better
pd.set_option('display.max_colwidth', -1)
projects_df.mjtheme_namecode.head(10)
```

```python
#1. Find the 10 countries with most projects
projects_df.groupby('countryname').size().sort_values(ascending=False).head(10)
```

```python
#2. Find the top 10 major project themes (using column 'mjtheme_namecode')
# load in data and use json_normalize to create tables from nested element using the id as an index (unique for each row)
data=json.load((open('data/world_bank_projects.json')))
mjtheme_namecode=json_normalize(data,'mjtheme_namecode',['id'])
```

```python
mjtheme_namecode.head()
```

```python
mjtheme_blanks = mjtheme_namecode[(mjtheme_namecode.name =="" )]
mjtheme_blanks.groupby('id').size().sort_values(ascending = False)
#note that there are no duplicate ids in the data set
```

```python
#here are my top 10 themes in descending order, head(11) is used because a large number of rows have a blank "name"
mjtheme_namecode.groupby('name').size().sort_values(ascending = False).head(11)
```

```python
#3. In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.
```

```python
#double check that every id possible has a theme name somewhere (so you can create a lookup from this dataframe)
themelkup = mjtheme_namecode[['code','name']]
themelkup.groupby(['code', 'name']).size()
```

```python
#drop blanks
themelkup2 = themelkup[themelkup['name'] != ""]
themelkup2
```

```python
#create a lookup table of theme id by name
themelkup3 = themelkup2.drop_duplicates()
themelkup3.set_index('code').sort_index()
```

```python
#merge lookup table onto old dataframe to fill in theme name
mjtheme_namecode2 = pd.merge(mjtheme_namecode, themelkup3, on='code' )
#result = pd.merge(left, right, on='key')
mjtheme_namecode2
```

```python
#clean up final dataframe, here are all my project id's with the name of the theme filled in
mjtheme_final = mjtheme_namecode2[['code','id', 'name_y']]
mjtheme_final
```
